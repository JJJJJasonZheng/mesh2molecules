### Collect every world in the name

```python
term = []
for i in df.name:
    temp = i.lower().split()
    for j in temp:
        if j not in term:
            term.append(j)
```

```python
len(term)
```

### Use definition to search (parts)

```python
result = []
count = 0
for i in term:
    count+=1
    query = "MATCH (n:ExternalOntology) WHERE n.synonym[0] CONTAINS '" + i \
    + "' WITH n MATCH (n) -[r]-> (m) RETURN m"
    with driver.session() as session:
        info = session.run(query)
        for item in info:
            temp = []
            temp.append(item.values()[0])
        result.append(temp)     
```
